# From bits to qubits

In existing computers, the smallest unit of information is the _bit_, that can be in one of two states, often denoted as $|0\rangle$ and $|1\rangle$.

In quantum information processing the analogue is the _qubit_ that is, similarly to the classical case, embedded in a two-level scheme. Since the qubit is quantum object it cannot only be in the states 0 and 1, it can also be in a _quantum superposition_ of the two states. The most general state $|\Psi \rangle$ of the qubit can be written as:

$\Psi = \alpha |0\rangle + \beta |1\rangle$

Due to the interpretation of the quantum state as a probability amplitude, the complex coefficients $\alpha$ and $\beta$ need to be normalized: $\alpha^2 + \beta^2 = 1$.

The two classical possibilities are at the southpole and northpole of the sphere. Every other state on the surface of the BLoch sphere is allowed by quantum mechanics, but cannot be implemented in a classical computer.

Since a classical system can have only two different states, the only operation that is allowed is a bit-flip, that will transfer the state $|0 \rangle$ into the state $|1 \rangle$
