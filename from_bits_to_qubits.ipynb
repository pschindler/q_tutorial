## Classical information

In existing computers, the smallest unit of information is the _bit_, that can be in one of two states, here denoted as $|0\rangle$ and $|1\rangle$. In this tutorial we will represent information as a point on a sphere. Classical information can only be points on the northpole and southpole.

```python
%matplotlib inline
from citizen_q import *
draw_ground()
```

## From bits to qubits



In quantum information processing, the smallest unit of information is the _qubit_ that is, similarly to the classical case, embedded in a two-level system. Since the qubit is a quantum object it can also be in a _quantum superposition_ of the two states. The most general state 
$|\Psi \rangle$
of the qubit can be written as:

$\Psi = \alpha |0\rangle + \beta |1\rangle$

Due to the interpretation of the quantum state as a probability amplitude, the complex coefficients $\alpha$ and $\beta$ need to be normalized: $\alpha^2 + \beta^2 = 1$. 

```python
draw_superposition()
```

The sphere above corresponds to the state 
$|\Psi_1\rangle  = 1/ \sqrt{2} \, (|0 \rangle + |1\rangle$ 
which is known as a superposition state that has no classical analogon.



## Creating a superposition
Let's create a superposition from a classical state. First, we will need a qubit in the ground state which is already defined in our helper library.

```python
psi = ground_state()
```

The state of a qubit can then be manipulated by performing _rotations_ around axes on the Bloch sphere. 
For example, we will rotate the ground state by 45° around the x-axis, which we can so by multiplying the state with the python command _rx(angle)_

```python
psi45 = rx(pi/4)*psi
```

The resulting state can then be displayed using the command show_sphere()

```python
show_sphere(psi45)
```
